```python
import os
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
folder="Datasets/"  # base folder having all datasets
files=os.listdir(folder)

for i,f in enumerate(files):
    if ".csv" not in f:
        files.pop(i)

r=0
df1=pd.read_csv(folder+files[r])   # read one dataset and remove it from the list to avoid redundancy
n=files[r].split('.')[0]
df1['Season']=n
#df1=df1.assign(Season=n)
print("len= ", len(df1))
files.pop(r)
df1.head()
```

```python
cols1=set(list(df1))   # get column names that are present in all datasets
for f in files:
    try:
        df2=pd.read_csv(folder+f)
        n=f.split('.')[0]
        #df2=df2.assign(Season=n)
        df2['Season']=n
        
        cols2=set(list(df2))
        cols1=cols1.intersection(cols2)
        print("F: ", f)
    except Exception as e:
        print(e)
        pass
cols1
```

```python
# append all datasets together based on the columns identified
cols=list(cols1)
for f in files:
    try:
        df2=pd.read_csv(folder+f)
        n=f.split('.')[0]
        df2['Season']=n
        
        df1=df1[cols]   # selecting only common columns from the data
        df2=df2[cols]
        
        df1=pd.concat([df1, df2])
        print("F: ", f)
    except Exception as e:
        print(list(df2))
        print(e)
        pass
print(df1.shape)
```

```python
df1=df1.dropna()  # drop rows having NANs if any
df1.head()
```

```python
df1.to_csv("final_data.csv")   # save data as final csv file
```
