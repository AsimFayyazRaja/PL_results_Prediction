```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
df=pd.read_csv('final_data.csv')  # read data
```

```python
df.head()   # display first 5 rows of the data
```

```python
df=df.sort_values('Date')   # sort on date
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
print(df.shape)
df.head()
```

```python
cols_to_keep=["Date","HomeTeam","AwayTeam","HS","AS","HST","AST","HC","AC","HTR","FTR","WHH","WHD",
             "WHA"]   # select only relevant columns from the data
df=df[cols_to_keep]
print(df.shape)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

le_team=LabelEncoder()

teams=df[['HomeTeam','AwayTeam']].values   # get all possible teams present in the data
teams=np.reshape(teams,(-1,1))
teams=teams.tolist()
```

```python
print(teams[:2])
```

```python
le_team.fit(np.array(teams).ravel())   # encode the teams from strings to numerical values
```

```python
df['HomeTeam']=le_team.transform(df['HomeTeam'].values)
df['AwayTeam']=le_team.transform(df['AwayTeam'].values)   # convert data to numerics
df.head()
```

```python
df['diff_shots']=df['HS']-df['AS']
df['diff_on_target']=df['HST']-df['AST']   # make two new features
df.head()
```

```python
le_half=LabelEncoder()   # encode column HTR
df['HTR']=le_half.fit_transform(df['HTR'])
df.head()
```

```python
df.to_csv('preprocessed_data.csv')   # save the preprocessed data for future use as CSV file
```

```python
to_drop=['Date','FTR']  # drop FTR and date from features
labels=df['FTR'].values   # take FTR as target values or labels
df=df.drop(to_drop,axis=1)
features=df.values
print(features.shape)
print(labels.shape)
```

```python
labs=[]
for l in labels:   # make probabilitic target values
    if l=='H':
        labs.append([1,0,0])   # home wins with 1 probability
    elif l=='D':
        labs.append([0,1,0])    # draw with 1 probability
    else:
        labs.append([0,0,1])    # away wins with 1 probability
labels=np.array(labs)
```

```python
import pickle

#saving data on disk for future use
with open('features.pkl','wb') as f:
    pickle.dump(features, f)

    
with open('le_team.pkl','wb') as f:
    pickle.dump(le_team, f)    
    
with open('labels.pkl','wb') as f:
    pickle.dump(labels, f)


# save with python2 compatibility as well
with open('features_py2.pkl','wb') as f:
    pickle.dump(features, f, protocol=2)

with open('labels_py2.pkl','wb') as f:
    pickle.dump(labels, f, protocol=2)
```
